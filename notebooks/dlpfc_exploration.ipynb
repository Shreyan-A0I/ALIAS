# Spatial DLPFC Data Exploration

This notebook explores the Liebar Institute's Spatial DLPFC dataset. Our goals are to:
1. Understand the `.h5ad` structure (genes, coordinates, spatial representation).
2. Understand the master metadata `.xlsx` file.
3. Ensure the samples present in the data align with the intended anterior sections from the mastersheet for our specific task.

```python
import scanpy as sc
import pandas as pd
import anndata as ad
import numpy as np
```

## 1. Inspecting the AnnData Object

```python
h5ad_path = "../spatialDLPFC_data.h5ad"
print(f"Loading data from {h5ad_path}...")

adata = sc.read_h5ad(h5ad_path)
print(adata)
```

```python
# Let's look at the spot metadata (`obs`)
adata.obs.head()
```

```python
# Checking for Spatial Coordinates
print("Available obsm keys (typically spatial coords are here):", adata.obsm.keys())

if 'spatial' in adata.obsm:
    print("\nFirst 5 Spatial coordinates:")
    print(adata.obsm['spatial'][:5])
```

## 2. Inspecting the Master Sheet Metadata

```python
xlsx_path = "../Visium_dlpfc_mastersheet.xlsx"
df_master = pd.read_excel(xlsx_path)

df_master.head()
```

## 3. Aligning and Validating Samples

Now we identify the unique samples in `adata` and match them to the `sample_id` (or similar) inside the master Excel sheet.

```python
# Finding the column that defines the sample in adata.obs
# Common names: 'sample_id', 'sample', 'batch'
sample_col = None
for col in adata.obs.columns:
    if 'sample' in col.lower() or 'batch' in col.lower():
        sample_col = col
        break

if sample_col:
    adata_samples = set(adata.obs[sample_col].unique())
    print(f"Found {len(adata_samples)} samples in AnnData under column '{sample_col}':")
    print(list(adata_samples)[:10], "..." if len(adata_samples) > 10 else "")
else:
    print("Could not automatically find a sample column in adata.obs")
    adata_samples = set()

# Finding the corresponding column in the Excel sheet
master_sample_col = None
for col in df_master.columns:
    if 'sample' in col.lower() or 'id' in col.lower():
        # Often the Excel sheet has 'sample_id' or 'sample_name'
        master_sample_col = col
        break

if master_sample_col:
    master_samples = set(df_master[master_sample_col].astype(str).unique())
    print(f"\nFound {len(master_samples)} samples in Master Sheet under column '{master_sample_col}':")
    
    overlap = adata_samples.intersection(master_samples)
    print(f"\n{len(overlap)} samples overlap between the AnnData object and the Master Sheet.")
    if len(overlap) > 0:
        print("Sample overlap:", list(overlap)[:5], "...")
else:
    print("Could not automatically find a sample column in the master sheet.")
```
