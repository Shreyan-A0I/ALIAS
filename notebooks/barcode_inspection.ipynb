# Inv-SHAF: Barcode Data Inspection

This notebook displays the raw contents of the spatial transcriptomics barcode mapping file (`tissue_positions_list.csv`). Note that depending on the Space Ranger version, this file might not have headers, but its standard format is:
`barcode`, `in_tissue`, `array_row`, `array_col`, `pxl_row_in_fullres`, `pxl_col_in_fullres`

```python
import pandas as pd
import os
import json

spatial_dir = "../data/spatial"
positions_path = os.path.join(spatial_dir, "tissue_positions_list.csv")

# Read raw CSV without assuming headers to see exactly what is in the file
df_raw = pd.read_csv(positions_path, header=None)
display(df_raw.head())
```

### Spatial Scalefactors

The `scalefactors_json.json` file contains the multipliers necessary to map the high-resolution or low-resolution tissue images to the spatial capture coordinates. It also dictates the full resolution spot diameter.

```python
scalefactors_path = os.path.join(spatial_dir, "scalefactors_json.json")

with open(scalefactors_path, "r") as f:
    scalefactors = json.load(f)

print(json.dumps(scalefactors, indent=4))
```
